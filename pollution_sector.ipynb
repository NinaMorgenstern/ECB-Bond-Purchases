# Emissions of the main air pollutants by sector group in the EEA-33

```python
df3 = pd.read_csv('share-of-eea-33-emissions-5.csv')
```

```python
df3
```

```python
df3['share:number'] = df3.groupby('pollutant_name:text', as_index=False)['emissions:number'].apply(lambda g: (g / g.sum())*100).values
```

```python
df3
```

```python
df3.groupby('sector_name_eea:text', as_index=False)['share:number'].mean().sort_values('share:number', ascending=False).head(4)
```
