```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data from AWS S3 (public access)
data = pd.read_csv("https://s3groupmorocco.s3.eu-central-1.amazonaws.com/data/Industry.csv", thousands=',')
data.head()
```

```python
data['Quartal'] = pd.to_datetime(data['Quartal'])
```

```python
data['Quartal'] = data['Quartal'].dt.to_period("Q")
```

```python
data
```

```python
sns.barplot(x = 'Quartal', y = 'Total Holdings', data = data)

plt.xlabel('Year quarterly')

plt.ylabel('Total holdings in Billions of Euro')

plt.title('ECB CSPP program')

plt.xticks(rotation = 30)
```
